```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Adam Material\PROJECTS\netflix_titles.csv\netflix_titles.csv")
```

```python
df.info()
```

```python
#relation for null values to every column
nulls_perc = df.isnull().sum() / len(df) * 100
sns.heatmap(nulls_perc.to_frame(), annot=True, cmap='coolwarm')
plt.show()
```

```python
df['director'].fillna("Unknown", inplace=True)    
df['cast'].fillna("Unknown", inplace=True)    
df['country'].fillna("Unknown", inplace=True)    
df['date_added'].fillna("Unknown", inplace=True)      
df['rating'].fillna("Unknown", inplace=True)    
df['duration'].fillna("Unknown", inplace=True)    
```

```python
df
```

```python
df ['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')
df
```

```python
df['movie_min_duration'] = df['duration'].str.extract(r'(\d+)(?=\s*min)').astype(float)
df['TV_show_Seasons'] = df['duration'].str.extract(r'(\d+)(?=\s*Seasons)').astype(float)
```

```python
df['movie_min_duration'].fillna(0, inplace=True)
df['TV_show_Seasons'].fillna(0, inplace=True)
```

```python
df
```

```python
df['type'].value_counts()
```

```python
category = ['Movie', 'TV Show']
Counts=[df['type'].value_counts()[0], df['type'].value_counts()[1]]
plt.pie(Counts, labels=category, autopct='%1.1f%%')
plt.show()
```

```python
top_10 = df['country'].value_counts()[0:11].drop('Unknown').reset_index()
top_10.columns = ['Country', 'Count']
```

```python

plt.bar(top_10['Country'], top_10['Count'], width=0.8)
plt.xticks(rotation=45, ha='right')
plt.title("TOP 10 Countries of (Movies, TV Shows)")
plt.show()
```

```python
most_year_add_after2015 = df['release_year'].value_counts()[0:5].reset_index()
most_year_add_after2015.columns = ['Year', 'Added']
```

```python
plt.bar(most_year_add_after2015['Year'], most_year_add_after2015['Added'])
plt.title("How many Movie and TV Show added after 2015")
plt.show()
```

```python
sns.boxplot(x=df['movie_min_duration'].drop(0.0))
plt.title("Distribution of Movie Durations")
plt.xlabel("Duration (minutes)")
plt.show()
```

```python
top_3_genres = df['listed_in'].value_counts()[0:3].reset_index()
top_3_genres.columns = ['Genre', 'Counts']
plt.pie(top_3_genres['Counts'], labels=top_3_genres['Genre'], autopct='%1.1f%%')
plt.show()
```
